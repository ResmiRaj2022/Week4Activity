# Count of female senior citizens there in the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


df = pd.read_excel (r'C:/Users/user/Desktop/Activity3/Telco-Customer-Churn.xlsx')
#print(df)



df3 =df[(df['gender']=='Female') & (df['SeniorCitizen']==1.0)]['SeniorCitizen'].count()
print("Count of Female senior Citizens", df3)

df.plot.box()
plt.boxplot([df3])
plt.show()








```
